#### **Data Cleaning**

```python
# Data Cleaning Action: Dropping the charging status, status, displayed comment, Kind, Geo point, charge slots
# slots, scheduled at, station type, address, subsription type, columns, cars
# Explanation:  They are not needed for analysis. Cars column is redundant with bluecar counter
columns = ['Charging Status', 'Status', 'Cars', 'Displayed comment', 'Address', 'Kind', 'Geo point', 'Charge Slots', 'Slots', 'Scheduled at', 'Station type', 'Subscription status']
df.drop(columns, axis = 1, inplace=True)

df.head(2)
```

```python
# Data Cleaning Action: Checking for missing values in the columns we have now
# Explanation: Missing values affect analysis therefore it is important to detect them
#No missing values
df.isnull().sum()
```

```python
# Data Cleaning Action: Checking for duplicates
# Explanation: Ensuring that rows are not repeated
#no duplicates found
df.duplicated().any()
```

```python
# Data Cleaning Action: standardizing column names
# Explanation: making all the column names the same case - lowercase, and removing the spaces between words
#
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns.values
```

#### **Creating combined date-time column**

```python
df.head(1)
```

```python
# creating date time column to aid in analysis
#creating a function that will join the year, month, day, hour and minute column values
def date(row):
  return (str(row['year']) + '-' + str(row['month']) + '-' + str(row['day']) + ' ' + str(row['hour']) + ':' + str(row['minute']) + ':' + '00')

#creating new column date applying the function above
df['date'] = df.apply((lambda row: date(row)), axis=1)

#coverting new date column to datetime
df['date'] = pd.to_datetime(df['date'])

#previewing dataframe with the new column
df.head(2)
```

```python
df['date_only'] = df['date'].dt.date
df['time_only'] = df['date'].dt.time
df.head(2)
```

##### Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
len(df[df["city"]== "Paris"])
```

```python
df.date_only.unique()
```

```python
#Identify the most popular hour of the
#day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

dfp = df[df["city"] == "Paris"]

dfp = pd.DataFrame(dfp.groupby(['public_name','date_only', 'time_only', 'hour'])['bluecar_counter'].sum())
dfp.head(10)
```

```python
len(dfp)
```

```python
#resetting the index to make the date, time and hour indexes columns
dfp.reset_index(level=[1,2,3], inplace = True)
dfp = dfp.assign(bluecar_diff=0)
dfp.head(10)
```

```python
for x in dfp.index.unique():
  #print(x)
  dfp.loc[x, "bluecar_diff"] = dfp.loc[[x]].bluecar_counter.diff().values
```

```python
dfp.tail(10)
```

```python
dfp.bluecar_diff.fillna("station's initial record", inplace=True)
```

```python
dfp.head(10)
```

```python
#picked up - negative value
#returned - positive
#0 or nan - nothing happened
def act(x):
  if type(x) != str:
    if x > 0:
      return "return"
    elif x < 0: 
      return "pick up"
    else:
      return "no return or pick up"
  else:
    return "no return or pick up"
dfp['action'] = dfp['bluecar_diff'].apply(lambda x: act(x))
```

```python
dfp.head(10)
```

```python
df_pick = dfp[dfp['action'] == "pick up"]
df_pick.groupby('hour')['action'].count().nlargest(1)
#most popular hour of the day for picking up a Bluecar in Paris over the month of April 2018
#is 21h
```

##### What is the most popular hour for returning cars? (considering all cities, and for Paris as well)

```python
#making a new dataframe that considers all cities, not just paris, by following the steps in the Paris question that encompasses

df1 = pd.DataFrame(df.groupby(['public_name','date_only', 'time_only', 'hour'])['bluecar_counter'].sum())
df1.head(10)


df1.reset_index(level=[1,2,3], inplace = True)
df1 = df1.assign(bluecar_diff=0)
df1.head(10)

for x in df1.index.unique():
  #print(x)
  df1.loc[x, "bluecar_diff"] = df1.loc[[x]].bluecar_counter.diff().values

df1.bluecar_diff.fillna("station's initial record", inplace=True)

df1['action'] = df1['bluecar_diff'].apply(lambda x: act(x))
```

```python
df1.tail(10)
```

```python
#the most popular hour for returning cars
#overall
df_ret = df1[df1['action'] == "return"]
df_ret.groupby('hour')['action'].count().nlargest(1)
```

```python
print(df_ret.groupby('hour')['action'].count().sum())
df_x = df1[df1['action'] == "pick up"]
print(df_x.groupby('hour')['action'].count().sum())
df_y = df1[df1['action'] == "no return or pick up"]
print(df_y.groupby('hour')['action'].count().sum())
```
